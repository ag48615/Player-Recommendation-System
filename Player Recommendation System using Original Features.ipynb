```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #Data Visualization 
import seaborn as sns  #Python library for Vidualization
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('overallPlayerRecord.csv')
df.head()
df.shape
```

```python
cl = pd.read_csv('clustered1.csv')
```

```python
kmeans = KMeans(n_clusters = 5)
```

```python
df['cluster2'] = kmeans.fit_predict(df[df.columns[2:]])
```

```python
cl = pd.merge(df, cl, on='pid', how='inner')
cl = cl[['pid', 'batRI', 'bowlRI', 'clusterRI', 'cluster2_y']]
cl.head()
```

```python
fig = plt.gcf()
fig.set_size_inches(12, 12)
plt.style.use('ggplot')
plt.scatter(cl.batRI, cl.bowlRI, c=cl.clusterRI, edgecolor = 'black',  s = 100, linewidth = .6, alpha = .75)
plt.xlabel("batting ranking index")
plt.ylabel("bowling ranking index")

# centers = kmeans.cluster_centers_
# plt.scatter(centers[:, 0], centers[:, 1], c = 'black', s=100, alpha=.75)

plt.show()
```

```python
fig = plt.gcf()
fig.set_size_inches(12, 12)
plt.style.use('ggplot')
plt.scatter(cl.batRI, cl.bowlRI, c=cl.cluster2_y, edgecolor = 'black', s = 100, linewidth = .6, alpha = .75)
plt.xlabel("batting ranking index")
plt.ylabel("bowling ranking index")

plt.show()
```
