```python
import pandas as pd
import numpy as np
```

```python
cl = pd.read_csv('clustered055.csv')
players = pd.read_csv('playerList.csv')
```

```python
# iplayers = players[players['cid'] != 12]
# iplayers.head()

```

```python
tplayers = pd.DataFrame.merge(cl, players, on='pid', how='inner')
```

```python
ipc = tplayers[tplayers['cid']  == 5]
ipc.cid in ipc.cid.values
ipc[ipc.clusterRI == 1]
# ipc.shape
```

```python
def selectPlayers(pclustered, team):
    pospl = pd.DataFrame()
    
    rankbat = list()
    rankbowl = list()
    
    pteam = pclustered[pclustered.cid == team]
    
#     print(pteam)
    
    ptbat = pteam[pteam.clusterRI == 0]
    ptbat = ptbat.sort_values(['batRI'], ascending=False).reset_index(drop=True)
#     print(ptbat)
    
    ptbowl = pteam[pteam.clusterRI == 4]    
    ptbowl = ptbowl.sort_values(['bowlRI'], ascending=False).reset_index(drop=True)
#     print(ptbowl)
    
    ptall = pteam[(pteam.clusterRI == 2 | (pteam.clusterRI == 3))]
    ptall.insert(2, 'rir', ptall.batRI/ptall.bowlRI)

    ptall = ptall.sort_values(['rir', 'batRI','bowlRI'], ascending = False).reset_index(drop=True)
#     print("ptall", ptbowl)
    
    for i in [0, 1, 2, 3, 4]:
        pospl = pospl.append(ptbat.iloc[i])
    
#     print("len bat - ", len(pospl))
    
    if len(ptbowl) >= 4:
        rng = 4
    else:
        rng = len(ptbowl)
        
#     print(rng)
    for b in range(0, rng):
#         print("BBB ", b)
        pospl = pospl.append(ptbowl.iloc[b])
        
    l = len(pospl)
#     print("len - ", l)
    for b in range(0, 11 - 6):
        pospl = pospl.append(ptall.iloc[b])
        
#     print(pospl)
        
#     batting ranking

    batpl = pospl.sort_values(['batRI'], ascending=False).reset_index(drop=True)
    
    for b in range(0, 11):
        rankbat.append((b+1, batpl.iloc[b]['name']))
        
#     bowling ranking

    bowlpl = pospl.sort_values(['bowlRI'], ascending=False).reset_index(drop=True)
    
    for b in range(0, 11):
        rankbowl.append((b+1, bowlpl.iloc[b]['name']))
        
    return (rankbat, rankbowl)
        
    
    
# selectPlayers(ipc, 5)
```

```python
ranking = selectPlayers(ipc, 5)

print('BATSMAN RANKING')
print('----------------------------------------------------')
for pl in ranking[0]:
    print('POSITION: ', pl[0], '- NAME: ', pl[1])
    
print()
print('BOWLER RANKING') 
print('----------------------------------------------------')
for pl in ranking[1]:
    print('POSITION: ', pl[0], '- NAME: ', pl[1])
```
